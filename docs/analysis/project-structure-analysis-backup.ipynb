## 2Ô∏è‚É£ Mapeamento dos Diret√≥rios Principais

Vamos mapear e visualizar a hierarquia dos diret√≥rios principais do projeto.

````python
class ProjectAnalyzer:
    """Analisador de projeto otimizado"""
    
    def __init__(self, project_root):
        self.project_root = project_root
        self.structure_data = []
        
    def safe_file_count(self, path):
        """Conta arquivos de forma segura"""
        try:
            items = os.listdir(path)
            files = [f for f in items if os.path.isfile(os.path.join(path, f))]
            return len(files)
        except (PermissionError, OSError):
            return 0
    
    def safe_dir_count(self, path):
        """Conta diretorios de forma segura"""
        try:
            items = os.listdir(path)
            dirs = [d for d in items if os.path.isdir(os.path.join(path, d)) 
                   and not d.startswith('.') and d != 'node_modules']
            return len(dirs)
        except (PermissionError, OSError):
            return 0
    
    def get_safe_size(self, path):
        """Calcula tamanho de forma segura"""
        try:
            total_size = 0
            for root, dirs, files in os.walk(path):
                dirs[:] = [d for d in dirs if not d.startswith('.') and d != 'node_modules']
                for file in files[:50]:  # Limitar arquivos
                    try:
                        file_path = os.path.join(root, file)
                        total_size += os.path.getsize(file_path)
                    except (PermissionError, OSError):
                        continue
            return total_size / 1024  # KB
        except Exception:
            return 0
    
    def analyze_directory_structure(self):
        """Analise da estrutura de diretorios"""
        print("Analisando estrutura de diretorios...")
        
        directory_descriptions = {
            'client': 'Frontend React/TypeScript',
            'server': 'Backend Node.js/Express',
            'docs': 'Documentacao',
            'configs': 'Configuracoes',
            'database': 'Banco de dados',
            'docker': 'Containerizacao',
            'scripts': 'Scripts utilitarios',
            'IA': 'Inteligencia Artificial',
            'nginx': 'Servidor Web',
            'data': 'Dados de exemplo'
        }
        
        try:
            for item in os.listdir(self.project_root):
                item_path = os.path.join(self.project_root, item)
                
                if os.path.isdir(item_path) and not item.startswith('.'):
                    print(f"   Analisando: {item}...")
                    
                    num_files = self.safe_file_count(item_path)
                    num_subdirs = self.safe_dir_count(item_path)
                    size_kb = self.get_safe_size(item_path)
                    
                    self.structure_data.append({
                        'Diretorio': item,
                        'Descricao': directory_descriptions.get(item, 'Diretorio padrao'),
                        'Num_Arquivos': num_files,
                        'Num_Subdiretorios': num_subdirs,
                        'Tamanho_KB': round(size_kb, 2)
                    })
                    
        except Exception as e:
            print(f"Erro na analise: {e}")
            
        return self.structure_data
    
    def create_structure_dataframe(self):
        """Cria DataFrame com a estrutura analisada"""
        if not self.structure_data:
            self.analyze_directory_structure()
            
        return pd.DataFrame(self.structure_data)

# Executar analise
analyzer = ProjectAnalyzer(PROJECT_ROOT)
structure_df = analyzer.create_structure_dataframe()

print("\nESTRUTURA DO PROJETO:")
print("=" * 60)
if not structure_df.empty:
    print(structure_df.to_string(index=False))
    
    # Estatisticas resumidas
    total_files_sum = structure_df['Num_Arquivos'].sum()
    total_dirs_sum = structure_df['Num_Subdiretorios'].sum() 
    total_size_sum = structure_df['Tamanho_KB'].sum()
    
    print(f"\nRESUMO ESTATISTICO:")
    print(f"   Total de arquivos: {total_files_sum}")
    print(f"   Total de subdiretorios: {total_dirs_sum}")
    print(f"   Tamanho total: {total_size_sum:.2f} KB")
    print(f"   Media arquivos por diretorio: {total_files_sum/len(structure_df):.1f}")
else:
    print("Nenhum dado estrutural foi coletado")
````

## 4Ô∏è‚É£ Estrutura do Frontend (Client)

Vamos analisar detalhadamente a estrutura do frontend React com TypeScript, incluindo componentes, p√°ginas, contextos e servi√ßos.

````python
class SimpleFileAnalyzer:
    """Analise simples de tipos de arquivo"""
    
    def __init__(self, project_root):
        self.project_root = project_root
        self.max_files = 1000
        
    def analyze_file_types(self):
        """Analise de tipos de arquivo"""
        print("Analisando tipos de arquivo...")
        
        file_extensions = Counter()
        language_files = defaultdict(list)
        total_analyzed = 0
        
        # Mapeamento simples
        language_mapping = {
            '.js': 'JavaScript',
            '.ts': 'TypeScript',
            '.jsx': 'React JSX', 
            '.tsx': 'React TSX',
            '.py': 'Python',
            '.html': 'HTML',
            '.css': 'CSS',
            '.json': 'JSON',
            '.md': 'Markdown',
            '.yml': 'YAML',
            '.sql': 'SQL'
        }
        
        try:
            for root, dirs, files in os.walk(self.project_root):
                # Filtrar diretorios
                dirs[:] = [d for d in dirs if not d.startswith('.') and d != 'node_modules']
                
                for file in files:
                    if total_analyzed >= self.max_files:
                        break
                        
                    ext = pathlib.Path(file).suffix.lower()
                    
                    if ext:
                        file_extensions[ext] += 1
                        language = language_mapping.get(ext, 'Outros')
                        language_files[language].append(file)
                    else:
                        file_extensions['sem_extensao'] += 1
                        language_files['Outros'].append(file)
                    
                    total_analyzed += 1
                
                if total_analyzed >= self.max_files:
                    break
                    
        except Exception as e:
            print(f"Erro na analise: {e}")
            
        return file_extensions, language_files, total_analyzed
    
    def create_report(self):
        """Criar relatorio"""
        file_extensions, language_files, total_analyzed = self.analyze_file_types()
        
        print(f"\nANALISE DE ARQUIVOS ({total_analyzed} arquivos)")
        print("=" * 50)
        
        # Top extensoes
        print("\nTOP 10 EXTENSOES:")
        for i, (ext, count) in enumerate(file_extensions.most_common(10), 1):
            percentage = (count / total_analyzed) * 100 if total_analyzed > 0 else 0
            print(f"   {i:2d}. {ext:12} -> {count:4d} arquivos ({percentage:5.1f}%)")
        
        # Por linguagem
        print(f"\nDISTRIBUICAO POR LINGUAGEM:")
        language_counts = {lang: len(files) for lang, files in language_files.items()}
        sorted_languages = sorted(language_counts.items(), key=lambda x: x[1], reverse=True)
        
        for lang, count in sorted_languages:
            percentage = (count / total_analyzed) * 100 if total_analyzed > 0 else 0
            print(f"   {lang:15} -> {count:4d} arquivos ({percentage:5.1f}%)")
        
        # DataFrame
        ext_data = []
        for ext, count in file_extensions.most_common(10):
            percentage = (count / total_analyzed) * 100 if total_analyzed > 0 else 0
            ext_data.append({
                'Extensao': ext,
                'Quantidade': count,
                'Percentual': round(percentage, 2)
            })
        
        return pd.DataFrame(ext_data), language_files

# Executar analise
file_analyzer = SimpleFileAnalyzer(PROJECT_ROOT)
file_extensions_df, language_distribution = file_analyzer.create_report()

# Exibir DataFrame
if not file_extensions_df.empty:
    print(f"\nTABELA DE EXTENSOES:")
    print(file_extensions_df.to_string(index=False))
````

## 6Ô∏è‚É£ Sistema de IA e Machine Learning

Vamos examinar o diret√≥rio de IA, analisar datasets, modelos e notebooks de treinamento para processamento inteligente de extratos banc√°rios.

````python
class SimpleDependencyAnalyzer:
    """Analise simples de dependencias"""
    
    def __init__(self, project_root):
        self.project_root = project_root
        
    def safe_json_load(self, file_path):
        """Carrega JSON seguro"""
        try:
            with open(file_path, 'r', encoding='utf-8') as f:
                return json.load(f)
        except Exception as e:
            print(f"Erro ao ler {file_path}: {e}")
            return None
    
    def find_package_files(self):
        """Encontra package.json"""
        package_files = []
        
        # Verificar principais
        main_locations = [
            ('Root', 'package.json'),
            ('Client', 'client/package.json'),
            ('Server', 'server/package.json')
        ]
        
        for name, rel_path in main_locations:
            full_path = os.path.join(self.project_root, rel_path)
            if os.path.exists(full_path):
                package_files.append((name, full_path))
        
        return package_files
    
    def analyze_dependencies(self):
        """Analise de dependencias"""
        print("Analisando dependencias...")
        
        package_files = self.find_package_files()
        results = {
            'packages': [],
            'dependencies': defaultdict(set),
            'total_deps': 0,
            'total_dev_deps': 0
        }
        
        for pkg_name, pkg_path in package_files:
            print(f"   Analisando: {pkg_name}")
            
            pkg_data = self.safe_json_load(pkg_path)
            if not pkg_data:
                continue
                
            # Info basica
            package_info = {
                'name': pkg_data.get('name', 'N/A'),
                'version': pkg_data.get('version', 'N/A'),
                'location': pkg_name
            }
            
            # Dependencias
            deps = pkg_data.get('dependencies', {})
            dev_deps = pkg_data.get('devDependencies', {})
            
            for dep in deps:
                results['dependencies'][dep].add(pkg_name)
            
            package_info['deps_count'] = len(deps)
            package_info['dev_deps_count'] = len(dev_deps)
            
            results['packages'].append(package_info)
            results['total_deps'] += len(deps)
            results['total_dev_deps'] += len(dev_deps)
        
        return results
    
    def create_report(self):
        """Criar relatorio"""
        results = self.analyze_dependencies()
        
        print("\nRELATORIO DE DEPENDENCIAS")
        print("=" * 50)
        
        print(f"\nRESUMO:")
        print(f"   Arquivos package.json: {len(results['packages'])}")
        print(f"   Total dependencias: {results['total_deps']}")
        print(f"   Total dev dependencies: {results['total_dev_deps']}")
        print(f"   Dependencias unicas: {len(results['dependencies'])}")
        
        # Detalhes por package
        if results['packages']:
            print(f"\nDETALHES:")
            for pkg in results['packages']:
                print(f"   {pkg['name']} v{pkg['version']}")
                print(f"      Local: {pkg['location']}")
                print(f"      Deps: {pkg['deps_count']}, Dev: {pkg['dev_deps_count']}")
        
        # Top dependencias
        if results['dependencies']:
            print(f"\nTOP 10 DEPENDENCIAS:")
            dep_usage = {dep: len(locations) for dep, locations in results['dependencies'].items()}
            top_deps = sorted(dep_usage.items(), key=lambda x: x[1], reverse=True)[:10]
            
            for i, (dep, count) in enumerate(top_deps, 1):
                print(f"   {i:2d}. {dep:25} -> usado em {count} package(s)")
        
        return results

# Executar analise
dep_analyzer = SimpleDependencyAnalyzer(PROJECT_ROOT)
dependency_results = dep_analyzer.create_report()
````

````python
class SimpleSecurityAnalyzer:
    """Analise simples de seguranca"""
    
    def __init__(self, project_root):
        self.project_root = project_root
        self.max_files = 100
        
    def basic_security_check(self):
        """Verificacao basica de seguranca"""
        print("Verificando seguranca basica...")
        
        security_issues = []
        files_checked = 0
        
        # Padroes simples de segredos
        secret_patterns = [
            ('password', 'senha'),
            ('api_key', 'chave de API'),
            ('secret', 'segredo'),
            ('token', 'token')
        ]
        
        try:
            for root, dirs, files in os.walk(self.project_root):
                dirs[:] = [d for d in dirs if not d.startswith('.') and d != 'node_modules']
                
                for file in files:
                    if files_checked >= self.max_files:
                        break
                        
                    if file.endswith(('.js', '.ts', '.env', '.json')):
                        file_path = os.path.join(root, file)
                        
                        try:
                            with open(file_path, 'r', encoding='utf-8', errors='ignore') as f:
                                content = f.read(1000)  # Primeiros 1KB
                            
                            for pattern, desc in secret_patterns:
                                if pattern in content.lower():
                                    security_issues.append({
                                        'file': os.path.relpath(file_path, self.project_root),
                                        'issue': desc,
                                        'type': 'potential_secret'
                                    })
                            
                            files_checked += 1
                            
                        except Exception:
                            continue
                
                if files_checked >= self.max_files:
                    break
                    
        except Exception as e:
            print(f"Erro no scan: {e}")
        
        return security_issues, files_checked
    
    def check_security_configs(self):
        """Verifica configuracoes de seguranca"""
        security_configs = {
            'gitignore_exists': os.path.exists(os.path.join(self.project_root, '.gitignore')),
            'env_example_exists': os.path.exists(os.path.join(self.project_root, '.env.example')),
            'dockerfile_exists': os.path.exists(os.path.join(self.project_root, 'Dockerfile')),
            'package_lock_exists': os.path.exists(os.path.join(self.project_root, 'package-lock.json'))
        }
        
        return security_configs
    
    def create_security_report(self):
        """Criar relatorio de seguranca"""
        issues, files_checked = self.basic_security_check()
        configs = self.check_security_configs()
        
        print("\nANALISE DE SEGURANCA")
        print("=" * 30)
        
        print(f"\nSCAN DE SEGREDOS:")
        print(f"   Arquivos verificados: {files_checked}")
        print(f"   Problemas encontrados: {len(issues)}")
        
        if issues:
            print(f"\n   PROBLEMAS DETECTADOS:")
            for issue in issues[:10]:  # Top 10
                print(f"      - {issue['file']}: {issue['issue']}")
        else:
            print("   Nenhum problema obvio detectado")
        
        print(f"\nCONFIGURACOES DE SEGURANCA:")
        config_labels = {
            'gitignore_exists': '.gitignore presente',
            'env_example_exists': '.env.example presente',
            'dockerfile_exists': 'Dockerfile presente',
            'package_lock_exists': 'package-lock.json presente'
        }
        
        for config, status in configs.items():
            label = config_labels.get(config, config)
            icon = "‚úì" if status else "‚úó"
            print(f"   {icon} {label}")
        
        # Score simples
        config_score = sum(configs.values()) / len(configs) * 100
        issue_penalty = min(len(issues) * 10, 50)
        total_score = max(config_score - issue_penalty, 0)
        
        print(f"\nSCORE DE SEGURANCA: {total_score:.1f}/100")
        
        if total_score >= 80:
            print("   Status: Boa seguranca")
        elif total_score >= 50:
            print("   Status: Seguranca basica")
        else:
            print("   Status: Melhorias necessarias")
        
        return {
            'issues': issues,
            'configs': configs,
            'security_score': total_score,
            'files_checked': files_checked
        }

# Executar analise
security_analyzer = SimpleSecurityAnalyzer(PROJECT_ROOT)
security_results = security_analyzer.create_security_report()
````

Sistema de corre√ß√£o automatizada para identificar e corrigir problemas estruturais, de configura√ß√£o e qualidade de c√≥digo.

````python
class SimpleAdvancedAnalyzer:
    """Analise avancada simplificada"""
    
    def __init__(self, project_root):
        self.project_root = project_root
        self.max_files = 50
        
    def analyze_code_metrics(self):
        """Metricas basicas de codigo"""
        print("Analisando metricas de codigo...")
        
        metrics = {
            'total_lines': 0,
            'code_files': 0,
            'functions_count': 0,
            'classes_count': 0
        }
        
        files_processed = 0
        
        try:
            for root, dirs, files in os.walk(self.project_root):
                dirs[:] = [d for d in dirs if not d.startswith('.') and d != 'node_modules']
                
                for file in files:
                    if files_processed >= self.max_files:
                        break
                        
                    if file.endswith(('.js', '.ts', '.jsx', '.tsx', '.py')):
                        file_path = os.path.join(root, file)
                        
                        try:
                            with open(file_path, 'r', encoding='utf-8', errors='ignore') as f:
                                lines = f.readlines()
                            
                            metrics['total_lines'] += len(lines)
                            metrics['code_files'] += 1
                            
                            # Contar funcoes e classes (simples)
                            for line in lines:
                                line = line.strip().lower()
                                if line.startswith('function ') or line.startswith('def '):
                                    metrics['functions_count'] += 1
                                elif line.startswith('class '):
                                    metrics['classes_count'] += 1
                            
                            files_processed += 1
                            
                        except Exception:
                            continue
                
                if files_processed >= self.max_files:
                    break
                    
        except Exception as e:
            print(f"Erro na analise: {e}")
        
        return metrics
    
    def create_advanced_report(self):
        """Relatorio avancado"""
        metrics = self.analyze_code_metrics()
        
        print("\nANALISE AVANCADA")
        print("=" * 25)
        
        print(f"\nMETRICAS DE CODIGO:")
        print(f"   Arquivos de codigo: {metrics['code_files']}")
        print(f"   Total de linhas: {metrics['total_lines']:,}")
        print(f"   Funcoes encontradas: {metrics['functions_count']}")
        print(f"   Classes encontradas: {metrics['classes_count']}")
        
        if metrics['code_files'] > 0:
            avg_lines = metrics['total_lines'] / metrics['code_files']
            print(f"   Media de linhas por arquivo: {avg_lines:.1f}")
        
        # Qualidade do codigo (simples)
        quality_score = 100
        
        if metrics['code_files'] > 0:
            lines_per_file = metrics['total_lines'] / metrics['code_files']
            if lines_per_file > 500:
                quality_score -= 20  # Arquivos muito grandes
            
            functions_per_file = metrics['functions_count'] / metrics['code_files']
            if functions_per_file < 2:
                quality_score -= 10  # Poucos funcoes
        
        print(f"\nQUALIDADE DO CODIGO: {quality_score}/100")
        
        if quality_score >= 80:
            print("   Status: Boa qualidade")
        elif quality_score >= 60:
            print("   Status: Qualidade aceitavel")
        else:
            print("   Status: Melhorias recomendadas")
        
        # Recomendacoes
        print(f"\nRECOMENDACOES:")
        if metrics['code_files'] == 0:
            print("   - Nenhum arquivo de codigo encontrado")
        else:
            if metrics['total_lines'] / metrics['code_files'] > 500:
                print("   - Considere quebrar arquivos grandes")
            if metrics['functions_count'] == 0:
                print("   - Considere modularizar o codigo")
            else:
                print("   - Codigo bem estruturado")
        
        return {
            'metrics': metrics,
            'quality_score': quality_score
        }

# Executar analise
advanced_analyzer = SimpleAdvancedAnalyzer(PROJECT_ROOT)
advanced_results = advanced_analyzer.create_advanced_report()
````

## üîç An√°lise de Depend√™ncias e Acoplamento
Vamos analisar as depend√™ncias entre m√≥dulos, detectar acoplamento alto e identificar depend√™ncias circulares.

````python
import ast
import json
import subprocess
from collections import defaultdict
import networkx as nx
import requests
from pathlib import Path
import warnings
warnings.filterwarnings('ignore')

class AdvancedProjectAnalyzer:
    """üîç Analisador Avan√ßado para Will Finance 5.0"""
    
    def __init__(self, project_path):
        self.project_path = Path(project_path)
        self.dependencies = defaultdict(list)
        self.import_graph = nx.DiGraph()
        
    def extract_imports_from_file(self, filepath):
        """Extrai imports de um arquivo Python"""
        try:
            with open(filepath, 'r', encoding='utf-8', errors='ignore') as file:
                content = file.read()
                tree = ast.parse(content)
            
            imports = []
            for node in ast.walk(tree):
                if isinstance(node, ast.Import):
                    imports.extend([alias.name for alias in node.names])
                elif isinstance(node, ast.ImportFrom) and node.module:
                    imports.append(node.module)
            
            return imports
        except Exception as e:
            return []
    
    def analyze_python_dependencies(self):
        """üìä Analisa depend√™ncias entre arquivos Python"""
        print("üîç Analisando depend√™ncias Python...")
        
        python_files = list(self.project_path.rglob('*.py'))
        python_files = [f for f in python_files if 'node_modules' not in str(f) and '.git' not in str(f)]
        
        print(f"üìù Encontrados {len(python_files)} arquivos Python")
        
        for py_file in python_files:
            imports = self.extract_imports_from_file(py_file)
            relative_path = str(py_file.relative_to(self.project_path))
            self.dependencies[relative_path] = imports
            
            # Adiciona ao grafo
            self.import_graph.add_node(relative_path)
            for imp in imports:
                self.import_graph.add_edge(relative_path, imp)
        
        return self.dependencies
    
    def find_circular_dependencies(self):
        """üîÑ Encontra depend√™ncias circulares"""
        try:
            cycles = list(nx.simple_cycles(self.import_graph))
            return cycles
        except:
            return []
    
    def calculate_coupling_metrics(self):
        """üìà Calcula m√©tricas de acoplamento"""
        coupling_metrics = {}
        
        for node in self.import_graph.nodes():
            if node.endswith('.py'):
                # Fan-out (quantos m√≥dulos este importa)
                fan_out = len(list(self.import_graph.successors(node)))
                
                # Fan-in (quantos m√≥dulos importam este)
                fan_in = len(list(self.import_graph.predecessors(node)))
                
                coupling_metrics[node] = {
                    'fan_out': fan_out,
                    'fan_in': fan_in,
                    'coupling_score': fan_out + fan_in
                }
        
        return coupling_metrics

# Inicializa o analisador avan√ßado
analyzer = AdvancedProjectAnalyzer(PROJECT_ROOT)

# Executa an√°lise de depend√™ncias
print("üöÄ INICIANDO AN√ÅLISE AVAN√áADA DE DEPEND√äNCIAS...")
dependencies = analyzer.analyze_python_dependencies()
coupling_metrics = analyzer.calculate_coupling_metrics()
circular_deps = analyzer.find_circular_dependencies()

print(f"\nüìä RESULTADOS:")
print(f"   ‚Ä¢ Total de arquivos Python analisados: {len(dependencies)}")
print(f"   ‚Ä¢ Depend√™ncias circulares encontradas: {len(circular_deps)}")
print(f"   ‚Ä¢ M√≥dulos com m√©tricas de acoplamento: {len(coupling_metrics)}")

# Top 5 m√≥dulos mais acoplados
if coupling_metrics:
    top_coupled = sorted(coupling_metrics.items(), key=lambda x: x[1]['coupling_score'], reverse=True)[:5]
    print(f"\nüîù TOP 5 M√ìDULOS MAIS ACOPLADOS:")
    for i, (module, metrics) in enumerate(top_coupled, 1):
        print(f"   {i}. {module.split('/')[-1]} (Score: {metrics['coupling_score']})")
        print(f"      ‚ÜóÔ∏è Importa: {metrics['fan_out']} | ‚ÜòÔ∏è √â importado por: {metrics['fan_in']}")

if circular_deps:
    print(f"\n‚ö†Ô∏è DEPEND√äNCIAS CIRCULARES DETECTADAS:")
    for i, cycle in enumerate(circular_deps[:3], 1):
        print(f"   {i}. {' ‚Üí '.join(cycle[:3])}...")
else:
    print(f"\n‚úÖ Nenhuma depend√™ncia circular detectada!")
````

## üß™ An√°lise de Cobertura de Testes
Verificando a presen√ßa e distribui√ß√£o de testes no projeto.

````python
def analyze_test_coverage():
    """üß™ Analisa cobertura de testes por m√≥dulo"""
    print("üîç Analisando cobertura de testes...")
    
    modules = ['client', 'server', 'IA', 'scripts', 'docs']
    test_coverage = {}
    test_patterns = ['test_*.py', '*.test.py', '*.spec.py', '*.test.js', '*.spec.js', '*.test.ts', '*.spec.ts']
    
    for module in modules:
        module_path = Path(PROJECT_ROOT) / module
        if not module_path.exists():
            test_coverage[module] = {'files': 0, 'patterns': [], 'status': 'M√≥dulo n√£o encontrado'}
            continue
        
        total_test_files = 0
        found_patterns = []
        
        for pattern in test_patterns:
            test_files = list(module_path.rglob(pattern))
            test_files = [f for f in test_files if 'node_modules' not in str(f)]
            total_test_files += len(test_files)
            if test_files:
                found_patterns.append(f"{pattern}: {len(test_files)}")
        
        # Verifica presen√ßa de diret√≥rios de teste
        test_dirs = []
        for test_dir in ['tests', 'test', '__tests__', 'spec']:
            if (module_path / test_dir).exists():
                test_dirs.append(test_dir)
        
        # Calcula arquivos de produ√ß√£o para estimativa de cobertura
        if module == 'client':
            prod_files = list(module_path.rglob('*.js')) + list(module_path.rglob('*.ts')) + list(module_path.rglob('*.jsx')) + list(module_path.rglob('*.tsx'))
        elif module == 'server':
            prod_files = list(module_path.rglob('*.js')) + list(module_path.rglob('*.ts')) + list(module_path.rglob('*.py'))
        else:
            prod_files = list(module_path.rglob('*.py')) + list(module_path.rglob('*.js')) + list(module_path.rglob('*.ts'))
        
        prod_files = [f for f in prod_files if 'node_modules' not in str(f) and '.git' not in str(f)]
        prod_files = [f for f in prod_files if not any(test_pattern.replace('*', '') in f.name for test_pattern in ['test_', '.test', '.spec'])]
        
        coverage_ratio = (total_test_files / len(prod_files) * 100) if len(prod_files) > 0 else 0
        
        status = "üî¥ Sem testes"
        if coverage_ratio > 50:
            status = "üü¢ Boa cobertura"
        elif coverage_ratio > 20:
            status = "üü° Cobertura m√©dia"
        elif coverage_ratio > 0:
            status = "üü† Poucos testes"
        
        test_coverage[module] = {
            'test_files': total_test_files,
            'prod_files': len(prod_files),
            'coverage_ratio': round(coverage_ratio, 1),
            'test_dirs': test_dirs,
            'patterns_found': found_patterns,
            'status': status
        }
    
    return test_coverage

def analyze_package_dependencies():
    """üì¶ Analisa depend√™ncias de pacotes"""
    print("üì¶ Analisando depend√™ncias de pacotes...")
    
    package_analysis = {}
    
    # Analisa package.json do cliente
    client_pkg = Path(PROJECT_ROOT) / "client" / "package.json"
    if client_pkg.exists():
        with open(client_pkg, 'r', encoding='utf-8') as f:
            data = json.load(f)
            deps = data.get('dependencies', {})
            dev_deps = data.get('devDependencies', {})
            
            package_analysis['frontend'] = {
                'dependencies': len(deps),
                'devDependencies': len(dev_deps),
                'total': len(deps) + len(dev_deps),
                'key_frameworks': [dep for dep in deps.keys() if dep in ['react', 'vue', 'angular', 'svelte', 'next']],
                'build_tools': [dep for dep in list(deps.keys()) + list(dev_deps.keys()) if dep in ['vite', 'webpack', 'parcel', 'rollup']],
                'ui_libraries': [dep for dep in deps.keys() if 'ui' in dep or dep in ['tailwindcss', 'bootstrap', 'material-ui', 'antd']]
            }
    
    # Analisa package.json do servidor
    server_pkg = Path(PROJECT_ROOT) / "server" / "package.json"
    if server_pkg.exists():
        with open(server_pkg, 'r', encoding='utf-8') as f:
            data = json.load(f)
            deps = data.get('dependencies', {})
            dev_deps = data.get('devDependencies', {})
            
            package_analysis['backend'] = {
                'dependencies': len(deps),
                'devDependencies': len(dev_deps),
                'total': len(deps) + len(dev_deps),
                'frameworks': [dep for dep in deps.keys() if dep in ['express', 'fastify', 'koa', 'nestjs']],
                'databases': [dep for dep in deps.keys() if dep in ['prisma', 'mongoose', 'typeorm', 'sequelize']],
                'auth': [dep for dep in deps.keys() if 'auth' in dep or dep in ['passport', 'jwt', 'bcrypt']]
            }
    
    # Analisa requirements.txt se existir
    requirements_file = Path(PROJECT_ROOT) / "requirements.txt"
    if requirements_file.exists():
        with open(requirements_file, 'r', encoding='utf-8') as f:
            lines = f.readlines()
            python_deps = [line.strip().split('==')[0] for line in lines if line.strip() and not line.startswith('#')]
            
            package_analysis['python'] = {
                'total': len(python_deps),
                'ml_libs': [dep for dep in python_deps if dep in ['tensorflow', 'pytorch', 'scikit-learn', 'pandas', 'numpy']],
                'web_frameworks': [dep for dep in python_deps if dep in ['flask', 'django', 'fastapi', 'tornado']],
                'data_processing': [dep for dep in python_deps if dep in ['pandas', 'numpy', 'matplotlib', 'seaborn', 'plotly']]
            }
    
    return package_analysis

# Executa an√°lises
test_coverage = analyze_test_coverage()
package_deps = analyze_package_dependencies()

# Exibe resultados de cobertura de testes
print("\nüß™ RESULTADOS DA AN√ÅLISE DE TESTES:")
print("=" * 50)
for module, info in test_coverage.items():
    print(f"\nüìÅ {module.upper()}:")
    print(f"   {info['status']}")
    print(f"   ‚Ä¢ Arquivos de teste: {info['test_files']}")
    print(f"   ‚Ä¢ Arquivos de produ√ß√£o: {info['prod_files']}")
    print(f"   ‚Ä¢ Cobertura estimada: {info['coverage_ratio']}%")
    if info['test_dirs']:
        print(f"   ‚Ä¢ Diret√≥rios de teste: {', '.join(info['test_dirs'])}")

# Exibe resultados de depend√™ncias
print(f"\nüì¶ AN√ÅLISE DE DEPEND√äNCIAS:")
print("=" * 50)
for platform, info in package_deps.items():
    print(f"\nüíª {platform.upper()}:")
    print(f"   ‚Ä¢ Total de depend√™ncias: {info['total']}")
    if 'dependencies' in info:
        print(f"   ‚Ä¢ Produ√ß√£o: {info['dependencies']} | Desenvolvimento: {info['devDependencies']}")
    
    # Mostra bibliotecas principais
    for category, libs in info.items():
        if isinstance(libs, list) and libs and category not in ['dependencies', 'devDependencies', 'total']:
            print(f"   ‚Ä¢ {category.replace('_', ' ').title()}: {', '.join(libs[:3])}{'...' if len(libs) > 3 else ''}")

# Recomenda√ß√µes baseadas na an√°lise
print(f"\nüí° RECOMENDA√á√ïES:")
print("=" * 50)

# Recomenda√ß√µes de testes
modules_without_tests = [module for module, info in test_coverage.items() if info.get('test_files', 0) == 0]
if modules_without_tests:
    print("üß™ TESTES:")
    for module in modules_without_tests:
        print(f"   ‚Ä¢ Implementar testes para o m√≥dulo '{module}'")

# Recomenda√ß√µes de depend√™ncias
if 'frontend' in package_deps and package_deps['frontend']['total'] > 100:
    print("üì¶ DEPEND√äNCIAS:")
    print("   ‚Ä¢ Frontend tem muitas depend√™ncias - considere otimiza√ß√£o")

if 'backend' in package_deps and not package_deps['backend']['auth']:
    print("   ‚Ä¢ Considere adicionar biblioteca de autentica√ß√£o no backend")
````
