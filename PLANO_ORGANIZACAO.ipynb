````python
# Gerando Script PowerShell para ReorganizaÃ§Ã£o AutomÃ¡tica

reorganization_script = '''
# ğŸ“ Script de ReorganizaÃ§Ã£o AutomÃ¡tica - Will Finance 5.0
# Reorganiza toda a estrutura do projeto seguindo melhores prÃ¡ticas

param(
    [switch]$DryRun = $false,
    [switch]$Force = $false
)

$ErrorActionPreference = "Stop"
$projectRoot = $PSScriptRoot

Write-Host "ğŸ”„ Iniciando ReorganizaÃ§Ã£o - Will Finance 5.0" -ForegroundColor Cyan
Write-Host "=" * 50 -ForegroundColor Gray

# =====================================================
# ğŸ› ï¸ FunÃ§Ãµes Auxiliares
# =====================================================

function Write-Status {
    param([string]$Message, [string]$Type = "INFO")
    switch ($Type) {
        "SUCCESS" { Write-Host "âœ… $Message" -ForegroundColor Green }
        "ERROR" { Write-Host "âŒ $Message" -ForegroundColor Red }
        "WARN" { Write-Host "âš ï¸ $Message" -ForegroundColor Yellow }
        "INFO" { Write-Host "â„¹ï¸ $Message" -ForegroundColor Cyan }
    }
}

function Move-SafeFile {
    param([string]$Source, [string]$Destination)
    
    if ($DryRun) {
        Write-Status "DRY RUN: Moveria $Source -> $Destination" "INFO"
        return
    }
    
    if (Test-Path $Source) {
        $destDir = Split-Path $Destination -Parent
        if (!(Test-Path $destDir)) {
            New-Item -Path $destDir -ItemType Directory -Force | Out-Null
        }
        
        if (Test-Path $Destination -and !$Force) {
            Write-Status "Destino jÃ¡ existe: $Destination" "WARN"
            return
        }
        
        Move-Item -Path $Source -Destination $Destination -Force
        Write-Status "Movido: $Source -> $Destination" "SUCCESS"
    } else {
        Write-Status "Arquivo nÃ£o encontrado: $Source" "WARN"
    }
}

# =====================================================
# ğŸ“‚ Etapa 1: Corrigir Nomenclatura de DiretÃ³rios
# =====================================================

Write-Host "`nğŸ“‚ Etapa 1: Corrigindo Nomenclatura..." -ForegroundColor Yellow

$directoryRenames = @{
    ".github" = ".github"
    ".husky_" = ".husky"  
    ".vscode" = ".vscode"
}

foreach ($oldName in $directoryRenames.Keys) {
    $newName = $directoryRenames[$oldName]
    $oldPath = Join-Path $projectRoot $oldName
    $newPath = Join-Path $projectRoot $newName
    
    if (Test-Path $oldPath) {
        if (Test-Path $newPath -and $oldName -ne $newName) {
            Write-Status "Mesclando $oldName com $newName existente..." "INFO"
            # Mesclar diretÃ³rios se necessÃ¡rio
        } else {
            Move-SafeFile $oldPath $newPath
        }
    }
}

# =====================================================
# ğŸ“ Etapa 2: Reorganizar Arquivos de ConfiguraÃ§Ã£o
# =====================================================

Write-Host "`nğŸ“ Etapa 2: Reorganizando ConfiguraÃ§Ãµes..." -ForegroundColor Yellow

$configMoves = @{
    "client.env" = "client/.env"
    "client.env.example" = "client/.env.example"
    "client.env.firebase" = "client/.env.firebase"
    "client.eslintrc.cjs" = "client/.eslintrc.cjs"
    "client.prettierrc.cjs" = "client/.prettierrc.cjs"
    "client.stylelintrc.cjs" = "client/.stylelintrc.cjs"
    "server.env" = "server/.env"
    "server.env.example" = "server/.env.example"
    "server.eslintrc.cjs" = "server/.eslintrc.cjs"
    "server.eslintrc.json" = "server/.eslintrc.json"
    "server.prettierrc" = "server/.prettierrc"
    "server.prettierrc.cjs" = "server/.prettierrc.cjs"
}

foreach ($source in $configMoves.Keys) {
    $destination = $configMoves[$source]
    $sourcePath = Join-Path $projectRoot $source
    $destPath = Join-Path $projectRoot $destination
    Move-SafeFile $sourcePath $destPath
}

# =====================================================
# ğŸ§ª Etapa 3: Unificar DiretÃ³rios de Testes
# =====================================================

Write-Host "`nğŸ§ª Etapa 3: Unificando Testes..." -ForegroundColor Yellow

# Mover testes do cliente
$clientTestSource = Join-Path $projectRoot "client/src_tests_"
$clientTestDest = Join-Path $projectRoot "client/src/tests"

if (Test-Path $clientTestSource) {
    if (!(Test-Path $clientTestDest)) {
        New-Item -Path $clientTestDest -ItemType Directory -Force | Out-Null
    }
    
    Get-ChildItem $clientTestSource -Recurse | ForEach-Object {
        $relativePath = $_.FullName.Substring($clientTestSource.Length + 1)
        $destFile = Join-Path $clientTestDest $relativePath
        Move-SafeFile $_.FullName $destFile
    }
    
    if (!$DryRun) {
        Remove-Item $clientTestSource -Recurse -Force
        Write-Status "Removido diretÃ³rio vazio: $clientTestSource" "SUCCESS"
    }
}

# Mover testes do servidor
$serverTestSource = Join-Path $projectRoot "server/src_tests_"
$serverTestDest = Join-Path $projectRoot "server/src/tests"

if (Test-Path $serverTestSource) {
    if (!(Test-Path $serverTestDest)) {
        New-Item -Path $serverTestDest -ItemType Directory -Force | Out-Null
    }
    
    Get-ChildItem $serverTestSource -Recurse | ForEach-Object {
        $relativePath = $_.FullName.Substring($serverTestSource.Length + 1)
        $destFile = Join-Path $serverTestDest $relativePath
        Move-SafeFile $_.FullName $destFile
    }
    
    if (!$DryRun) {
        Remove-Item $serverTestSource -Recurse -Force
        Write-Status "Removido diretÃ³rio vazio: $serverTestSource" "SUCCESS"
    }
}

Write-Host "`nğŸ‰ ReorganizaÃ§Ã£o ConcluÃ­da!" -ForegroundColor Green
Write-Host "Verifique os resultados e execute novamente sem -DryRun se tudo estiver correto." -ForegroundColor Cyan
'''

# Salvar o script
with open("reorganize-structure.ps1", "w", encoding="utf-8") as f:
    f.write(reorganization_script)

print("âœ… Script de reorganizaÃ§Ã£o criado: reorganize-structure.ps1")
print("ğŸ“‹ Para executar:")
print("   â€¢ Teste primeiro: .\\reorganize-structure.ps1 -DryRun")
print("   â€¢ Execute: .\\reorganize-structure.ps1 -Force")
````

### ğŸ“„ Arquivos
